```python
filename = '/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/Data Preprocessing/movie_dataset.csv'
movies_dataset_df = pd.read_csv('/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/Data Preprocessing/movie_dataset.csv')
movies_dataset_df.head()
```

```python
# 개인 맞춤 장르요소 새로운 행추가
# 213 -> 'Action Action Action Adventure Adventure Thriller Thriller Sci-Fi Sci-Fi Drama'
# 475 -> 'Drama Drama Drama Action Action Action Thriller Thriller Adventure Sci-Fi'
# 288 -> 'Comedy Comedy Comedy Action Action Action Drama Drama Adventure Adventure Romance'
# 64 -> 'user64_taste' ,'genres' : 'Drama Drama Drama Action Action Adventure Adventure Thriller Comedy'
new_data = {'title': 'user64_taste' ,'genres' : 'Drama Drama Drama Action Action Adventure Adventure Thriller Comedy'}
movies_dataset_df = movies_dataset_df.append(new_data, ignore_index=True)
```

```python
movies_dataset_df.tail(1)
```

```python
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_dataset_df['genres'])
print(genre_mat.shape)
```

```python
genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

```python
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

```python
percentile = 0.6
m = movies_dataset_df['naver_user_participate'].quantile(percentile)
C = movies_dataset_df['naver_user_rate'].mean()

def weighted_rating_average(record):
    v = record['naver_user_participate']
    R = record['naver_user_rate']
    
    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )   

movies_dataset_df['weighted_rate'] = movies_dataset_df.apply(weighted_rating_average, axis=1) 
```

```python
movies_dataset_df[['title','naver_user_rate','weighted_rate','naver_user_participate']].sort_values('weighted_rate', ascending=False)[:10]
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
  title_movie = df[df['title'] == title_name]
  title_index = title_movie.index.values
    
  similar_indexes = sorted_ind[title_index, :(top_n*2)]
  similar_indexes = similar_indexes.reshape(-1)

  similar_indexes = similar_indexes[similar_indexes != title_index]

  result_df = df.iloc[similar_indexes].sort_values('weighted_rate', ascending=False)[:top_n]
  result_df = result_df[['movieId', 'title', 'title_kor', 'genres', 'poster_url']]
  result_df.reset_index(drop=True, inplace=True)
  result_df.reset_index(drop=False, inplace=True)
  result_df.rename(columns={'index':'rank'}, inplace=True)
  result_df['rank'] = result_df['rank'] + 1
    
  return result_df
```

```python
recomm_movies = find_sim_movie(movies_dataset_df, genre_sim_sorted_ind, 'user64_taste',5)
recomm_movies['movieId'] = recomm_movies['movieId'].astype('int')
recomm_movies
```

```python
recomm_movies.to_csv("contents_based_filtering_list(genre).csv", mode='w', index=False)
```
