```python
filename = '/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/Data Preprocessing/movie_dataset.csv'
movie_df = pd.read_csv('/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/Data Preprocessing/movie_dataset.csv')
movie_df.head(3)
```

```python
rating_df = rating_df[['userId', 'movieId', 'rating']]
rating_matrix = rating_df.pivot_table('rating', index='userId', columns='movieId')
rating_matrix.head()
```

```python
rating_title = pd.merge(rating_df, movie_df, on='movieId')
rating_matrix = rating_title.pivot_table('rating', index='userId', columns='title')
rating_matrix = rating_matrix.fillna(0)
rating_matrix.head()
```

```python
rating_matrix_T = rating_matrix.transpose()
rating_matrix_T.head()
```

```python
item_sim = cosine_similarity(rating_matrix_T, rating_matrix_T)
item_sim_df = pd.DataFrame(data=item_sim, index=rating_matrix.columns ,columns=rating_matrix.columns)

print(item_sim_df.shape)
item_sim_df.head()
```

```python
item_sim_df["12 Years a Slave"].sort_values(ascending=False)[1:6]
```

```python
def predict_rating(ratings_arr, item_sim_arr):
  ratings_pred = ratings_arr.dot(item_sim_arr)/ np.array([np.abs(item_sim_arr).sum(axis=1)]) 
  return ratings_pred 

ratings_pred = predict_rating(rating_matrix.values , item_sim_df.values) 
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index= rating_matrix.index, columns = rating_matrix.columns) 
ratings_pred_matrix.head()
```

```python
def get_mse(pred, actual): 
  pred = pred[actual.nonzero()].flatten() 
  actual = actual[actual.nonzero()].flatten() 
  return mean_squared_error(pred, actual) 
  
print('아이템 기반 모든 인접 이웃 MSE: ', get_mse(ratings_pred, rating_matrix.values ))
```

```python
def predict_rating_topsim(ratings_arr, item_sim_arr, n=20): 
  pred = np.zeros(ratings_arr.shape) 
  for col in range(ratings_arr.shape[1]): 
    top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]] 
    for row in range(ratings_arr.shape[0]): 
      pred[row, col] = item_sim_arr[col, :][top_n_items].dot(ratings_arr[row, :][top_n_items].T) 
      pred[row, col] /= np.sum(np.abs(item_sim_arr[col, :][top_n_items])) 
  return pred 
  
ratings_pred = predict_rating_topsim(rating_matrix.values , item_sim_df.values, n=20) 
print('아이템 기반 인접 TOP-20 이웃 MSE: ', get_mse(ratings_pred, rating_matrix.values )) 

ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index= rating_matrix.index, columns = rating_matrix.columns)
```

```python
user_rating_id = rating_matrix.loc[9, :] 
user_rating_id[ user_rating_id > 0].sort_values(ascending=False)[:10]
```

```python
def get_unseen_movies(ratings_matrix, userId):  
  user_rating = ratings_matrix.loc[userId,:] 

  already_seen = user_rating[ user_rating > 0].index.tolist() 
  
  movies_list = ratings_matrix.columns.tolist() 
  
  unseen_list = [ movie for movie in movies_list if movie not in already_seen] 
  
  return unseen_list 

def recomm_movie_by_userid(pred_df, userId, unseen_list, top_n): 
  recomm_movies = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n] 
  return recomm_movies 

unseen_list = get_unseen_movies(rating_matrix, 64) 
recomm_movies = recomm_movie_by_userid(ratings_pred_matrix, 64, unseen_list, top_n=5) 
 
result_df = pd.DataFrame(data=recomm_movies.values,index=recomm_movies.index,columns=['pred_score'])
result_df
```

```python
result_df.reset_index(drop=False, inplace=True)
result_df.reset_index(drop=False, inplace=True)
result_df.rename(columns={'index':'rank'}, inplace=True)
result_df['rank'] = result_df['rank'] + 1
result_df
```

```python
recomm_movies = pd.merge(result_df, movie_df, how = 'inner', on=['title'])
recomm_movies = recomm_movies[['rank', 'movieId', 'title', 'title_kor', 'pred_score', 'poster_url']]
recomm_movies
```

```python
recomm_movies.to_csv("nearest neighbor collaborative filtering.csv", mode='w', index=False)
```
