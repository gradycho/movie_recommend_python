```python
# 45.5k movie_lens_data keywords load
filename = '/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/Data Preprocessing/movie_lens_data/45.5k/keywords.csv'
keywords_df = pd.read_csv('/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/Data Preprocessing/movie_lens_data/45.5k/keywords.csv')
keywords_df.head(2)
```

```python
# keywords_df, credits_df, movies_metadata_df의 ['id']컬럼 안맞는 데이터 드랍하고 int로 형변환
keywords_df['id'] = keywords_df['id'].astype('int')
credits_df['id'] = credits_df['id'].astype('int')
movies_metadata_df = movies_metadata_df.drop([19730, 29503, 35587])
movies_metadata_df['id'] = movies_metadata_df['id'].astype('int')
```

```python
# movies_metadata_df에 credits_df, keywords_df를 ['id']컬럼 기준으로 merge
movies_metadata_df = movies_metadata_df.merge(credits_df, on='id')
movies_metadata_df = movies_metadata_df.merge(keywords_df, on='id')
```

```python
# movies_metadata_df의 ['id']컬럼 결측치 확인
movies_metadata_df['id'].isna().sum()
```

```python
# movies_metadata_df['cast'], movies_metadata_df['crew'], movies_metadata_df['keywords']를 literal_eval로 str -> list 형변환
movies_metadata_df['cast'] = movies_metadata_df['cast'].apply(literal_eval)
movies_metadata_df['crew'] = movies_metadata_df['crew'].apply(literal_eval)
movies_metadata_df['keywords'] = movies_metadata_df['keywords'].apply(literal_eval)
```

```python
# director 값만 받아오는 함수 제작
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# ['director'], ['cast'], ['keywords'] 값만 뽑아오기
movies_metadata_df['director'] = movies_metadata_df['crew'].apply(get_director)

movies_metadata_df['cast'] = movies_metadata_df['cast'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
movies_metadata_df['cast'] = movies_metadata_df['cast'].apply(lambda x: x[:3] if len(x) >=3 else x)
movies_metadata_df['cast'] = movies_metadata_df['cast'].apply(lambda x: ','.join(x))

movies_metadata_df['keywords'] = movies_metadata_df['keywords'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
movies_metadata_df['keywords'] = movies_metadata_df['keywords'].apply(lambda x: ','.join(x))
```

```python
print(movies_metadata_df.shape)
movies_metadata_df.head()
```

```python
# movies_metadata_df['title']에서 (xxxx) 연도 제거, 글자 전부 소문자로 변환, 모든 공백 제거, 'only_title'로 새 컬럼에 추가
movies_metadata_df['only_title'] = movies_metadata_df['title'].str.lower()
movies_metadata_df['only_title'] = movies_metadata_df['only_title'].str.replace(" ", "")
movies_metadata_df = movies_metadata_df.dropna(subset = ['only_title'])
print(movies_metadata_df[['only_title']].isna().sum())

# movies_metadata_df['title']에서 (xxxx) 괄호 안 연도만 따로 추출해서 'openYear'로 새 컬럼에 추가
movies_metadata_df = movies_metadata_df.dropna(subset = ['release_date'])
movies_metadata_df['openYear'] = movies_metadata_df['release_date'].str[:4]
movies_metadata_df = movies_metadata_df.astype({'openYear' : int})
print(movies_metadata_df[['openYear']].isna().sum())

# movies_metadata_df['director'] 결측치 제거
movies_metadata_df = movies_metadata_df.dropna(subset = ['director'])
print(movies_metadata_df[['director']].isna().sum())

# movies_metadata_df['cast'] 결측치 제거
movies_metadata_df = movies_metadata_df.dropna(subset = ['cast'])
print(movies_metadata_df[['cast']].isna().sum())
```

```python
movies_metadata_cd_df = movies_metadata_df[['only_title', 'openYear', 'director', 'cast', 'keywords']]
movies_metadata_cd_df
```

```python
ml_movies_df = pd.merge(ml25m_movies_df, movies_metadata_cd_df, how = 'inner', on=['only_title', 'openYear'])
ml_movies_df
```

```python
# naver_movies_df['title_forn']에서 글자 전부 소문자로 변환, 모든 공백 제거, 'only_title'로 새 컬럼에 추가, 결측치 제거
naver_movies_df['only_title'] = naver_movies_df['title_forn'].str.lower()
naver_movies_df['only_title'] = naver_movies_df['only_title'].str.replace(" ", "")
naver_movies_df = naver_movies_df.dropna(subset = ['only_title'])

print(naver_movies_df[['only_title']].isna().sum())
naver_movies_df[['only_title']].head()
```

```python
print(ml_movies_df.shape)
print(naver_movies_df.shape)
```

```python
# ['only_title', 'openYear'] 두 컬럼을 기준으로 ml25m_movies_df와 naver_movies_df를 교집합('inner') 병합(merge)
# ['openYear'] 개봉연도 컬럼 차이를 (for문으로) 하나씩 더해가는 보정을 취하면서 일치하는 것을 매칭하고, same_movies_df 데이트 프레임에 추가
def merge_movies_data(df1, df2, n):
  same_movies_df = pd.DataFrame()

  for i in range(n):
    ml25m_movies_openYear_correction_df = df1.copy()
    ml25m_movies_openYear_correction_df['openYear'] = ml25m_movies_openYear_correction_df['openYear'] + i
    add_same_movies_df = pd.merge(ml25m_movies_openYear_correction_df, df2, how = 'inner', on=['only_title', 'openYear'])
    same_movies_df = pd.concat([same_movies_df, add_same_movies_df], axis = 0)
  
  same_movies_df = same_movies_df.drop_duplicates(['only_title', 'openYear'])
  same_movies_df = same_movies_df.sort_values('movieId')
  same_movies_df.reset_index(drop=True, inplace=True)

  # 컬럼명과 컬럼 순서 변경, 불필요한 컬럼 드랍, int로 자료형 변환, 문자열 구분자 변경, reset_index
  # ['only_title', 'openYear'] 행의 중복 제거(일치하는 행들의 가장 앞 순서 행만 남기고 나머지 드랍)
  same_movies_df.rename(columns={'code':'naver_code', 'title_forn':'title', 'title_y':'title_kor', 'keywords' : 'keyword', 'directors':'director_kor', 'actors':'cast_kor', 'openDate':'release_date', 'runningTime':'running_time', 
                              'netizens_rate':'naver_user_rate', 'netizens_participate':'naver_user_participate', 'critics_rate':'naver_critics_rate', 'critics_participate':'naver_critics_participate'}, inplace=True)
  same_movies_df.drop(['title_x'], axis=1, inplace=True)
  same_movies_df = same_movies_df.drop(['only_title', 'openYear'], axis=1)
  same_movies_df['genres'] = same_movies_df['genres'].str.replace("|", " ")
  same_movies_df['naver_user_participate'] = same_movies_df['naver_user_participate'].str.replace(",", "")
  same_movies_df = same_movies_df.astype({'naver_user_participate' : int})
  same_movies_df = same_movies_df.astype({'naver_critics_participate' : int})
  same_movies_df.reset_index(drop=True, inplace=True)
  same_movies_df = same_movies_df[['movieId', 'naver_code', 'title', 'title_kor', 'genres', 'keyword', 'director', 'cast', 'director_kor', 'cast_kor', 
                               'naver_user_rate', 'naver_user_participate', 'naver_critics_rate', 'naver_critics_participate', 'nation', 'release_date', 'running_time', 'age', 'story', 'poster_url', 'url']]

  return same_movies_df
```

```python
movie_db_df = merge_movies_data(ml_movies_df, naver_movies_df, 10)
movie_db_df.sample(2)
```

```python
# 결과 데이터프레임 shape과 type 확인
print(movie_db_df.shape)
print(movie_db_df.dtypes)
```
