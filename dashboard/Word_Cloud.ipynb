```python
user_rating_df = rating_df[(rating_df.userId == 288) & (rating_df.rating >= 3.5)]
user_rating_df.sort_values('rating')
```

```python
user_rating_df.value_counts('rating')
```

```python
user_rating_df[['movieId']]
```

```python
user_keyword_df = pd.merge(movie_df[['movieId', 'keyword']], user_rating_df[['userId', 'movieId', 'rating']], how = 'inner', on='movieId')
user_keyword_df = user_keyword_df.drop_duplicates(['movieId'], keep = 'first')
user_keyword_df = user_keyword_df.dropna(subset = ['keyword'])
user_keyword_df = user_keyword_df[['userId', 'movieId', 'rating', 'keyword']].sort_values('rating').reset_index(drop=True)
user_keyword_df.keyword = user_keyword_df.keyword.str.lower()
user_keyword_df.keyword = user_keyword_df.keyword.str.replace(" ", "")
user_keyword_df.keyword = user_keyword_df.keyword.str.replace(",", " ")
user_keyword_df
```

```python
user_keyword_df['keyword'][user_keyword_df['rating'] == 3.5] = user_keyword_df['keyword'][user_keyword_df['rating'] == 3.5].apply(lambda x: [x, x])
user_keyword_df['keyword'][user_keyword_df['rating'] == 3.5] = user_keyword_df['keyword'][user_keyword_df['rating'] == 3.5].apply(lambda x: ' '.join(x))

user_keyword_df['keyword'][user_keyword_df['rating'] == 4] = user_keyword_df['keyword'][user_keyword_df['rating'] == 4].apply(lambda x: [x, x, x])
user_keyword_df['keyword'][user_keyword_df['rating'] == 4] = user_keyword_df['keyword'][user_keyword_df['rating'] == 4].apply(lambda x: ' '.join(x))

user_keyword_df['keyword'][user_keyword_df['rating'] == 4.5] = user_keyword_df['keyword'][user_keyword_df['rating'] == 4.5].apply(lambda x: [x, x, x, x])
user_keyword_df['keyword'][user_keyword_df['rating'] == 4.5] = user_keyword_df['keyword'][user_keyword_df['rating'] == 4.5].apply(lambda x: ' '.join(x))

user_keyword_df['keyword'][user_keyword_df['rating'] == 5] = user_keyword_df['keyword'][user_keyword_df['rating'] == 5].apply(lambda x: [x, x, x, x, x])
user_keyword_df['keyword'][user_keyword_df['rating'] == 5] = user_keyword_df['keyword'][user_keyword_df['rating'] == 5].apply(lambda x: ' '.join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer()
m = count_vec.fit_transform(user_keyword_df.keyword)
m.toarray()
```

```python
keyword_counts = np.sum(m.toarray(), axis=0)
# len(keyword_counts)
keyword_counts_df = pd.DataFrame(keyword_counts)
keyword_counts_df
```

```python
keyword_index = count_vec.vocabulary_
# len(keyword_index)
keyword_index = dict(zip(keyword_index.values(),keyword_index.keys()))
keyword_index_df = pd.DataFrame([keyword_index])
keyword_index_df = keyword_index_df.transpose().sort_index()
keyword_index_df
```

```python
keyword_info_df = pd.concat([keyword_index_df, keyword_counts_df], axis = 1)
keyword_info_df.columns = ['keyword', 'count']
keyword_info_df = keyword_info_df.sort_values('count', ascending = False).reset_index(drop=True)
keyword_info_df = keyword_info_df[:50]
keyword_info_df.to_csv("keyword_info.csv", mode='w', index=False)
keyword_info_df
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image

word_freq = keyword_info_df['count'].to_dict()

im = Image.open('/content/drive/MyDrive/bigdata_analysis_final_project - movie recommender system(영화 추천 시스템)/dash board/projector3.png') # 이미지 파일 읽어오기
mask_arr = np.array(im) # 픽셀 값 배열 형태 변환

wordcloud = WordCloud(background_color ='white', colormap='winter', width = 1200, height = 800, random_state = 43, mask = mask_arr, prefer_horizontal = True).generate_from_frequencies(word_freq)

plt.figure(figsize = (12, 8))  
plt.imshow(wordcloud)
plt.title("movie taste keywords", size = 20)

plt.axis('off')

plt.savefig('./kwc50.png')
plt.show()
```
